```python
import numpy as np
from Data import Numbers
from nn import Network
```

```python
# Data structure for accessing mnist data
n = Numbers("./Data/mnist.pkl.gz")
```

```python
# Initial network for max pooling to 28 inputs instead of 784
X = np.array(n.train[0])
net1 = Network([28,28,28,10])
X.shape
```

```python
for j in range (28):
    temp = X[1][28*j:28*j+28]
    print("%s:\tlen: %d" % (str(temp),len(temp)))
```

```python
#Max pooled input data
X1 = np.zeros((50000,28))
for i in range(50000):
    for j in range(28):
        X1[i][j] = sum(X[i][28*j:28*j+28])
        
X1
```

```python
# Adjust labels to fit the 10x1 output requirement
y1 = np.zeros((50000,10))
for i in range(50000):
    y1[i][n.train[1][i]] = 1
    
y1.shape
```

```python
X1[1]
```

```python
net1.SGD_train(zip(X1,y1),10,0.1)
```

```python
#Test points using training data
print("Evaluating on test points")
randomTests = [[np.array(X1[i]).reshape(len(X1[i]),1),np.array(y1[i]).reshape(len(y1[i]),1)] for i in np.random.randint(0,50000,size=20)]
randomTests
```

```python
net1.evaluate(randomTests)
```

```python
np.array(randomTests[0][0]).shape
```

```python
net1.forward_prop(np.array(randomTests[0][0]).reshape(28,1))
```
